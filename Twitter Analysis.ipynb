```python
#Creating a single string from of our list of text into forms of tokens inorder to generate a wordcloud in further steps

def create_wc_string(tweettext):          #creating a function to clean and pre-process the tweets for further analysis
    raw_string = ''.join(str(v) for v in tweettext)                        #creating a string of the tweet
    no_links = re.sub(r'http\S+', '', raw_string)                          #replacing the links
    no_unicode = re.sub(r"\\[a-z][a-z]?[0-9]+", '', no_links)              #removing unicodes/emojis
    no_special_characters = re.sub('[^A-Za-z ]+', '', no_unicode)          #removing special characters 
    words = no_special_characters.split(" ")                               #splitting the words into tokens                       
    words = [w for w in words if len(w) > 2]                               #removing words with less than 2 characters
    words = [w.lower() for w in words]                                  
    words = [w for w in words if w not in STOPWORDS]
    return words
```

```python
#Applying the above created function on both the Twitter files collected 

tweets_bernie = create_wc_string(df_bernietweets['Text'])             
tweets1 = create_wc_string(df['Text'])
```

## Creating and comparing the WordClouds from the collected data of tweets corresponding to the terms "trump" and "bernie sanders"

```python
#creating a function that'd initialize wordclouds and prints a single string of tokenized words

def create_wc(tweetlist):
    wc = WordCloud(background_color="white", max_words=2000, height = 200, width = 200)
    clean_string = ','.join(tweetlist)                                   #creates a single string of the tokenized words                                           
    return wc.generate(clean_string)
```

```python
wordcloud_bernie = create_wc(tweets_bernie)
wordcloud_trump = create_wc(tweets1)
```

```python
#Creating a fucntion to display the wordclouds created above

def display_wc(wc, titlestr):
    f = plt.figure(figsize=(15,15))    #initializing the size of the ordcloud
    f.add_subplot(1,2, 2)                  
    plt.imshow(wc, interpolation='bilinear')    #to make the text less pixelated
    plt.title(titlestr, size=40)
    plt.axis("off")
    plt.show()
```

```python
display_wc(wordcloud_bernie, 'Bernie Wordcloud')
display_wc(wordcloud_trump, 'Trump Wordcloud')
```

```python
#After having dropped the stopwords, removed links, special characters and any other pre-processing required
#Now, counting the frequency of the words present in the tokenized version of all the tweets

#using the collections package and the method counter() inorder to count the occurrence of all the words in the tweets
#creating a function for this purpose

def counting(words):
    freq = collections.Counter(words)
    return freq
```

```python
#frequency of words in Bernie related tweets

tweetfreq_bernie = counting(tweets_bernie)
tweetfreq_bernie.most_common(10)            #selecting the top 10 most used words
```

# Sentiment Analysis and Comparison

```python
#Performing sentiment analysis and comparing

tweet_sentiments_bernie = df_bernietweets['Text']
tweet_sentiments_trump = df['Text']

tweet_sentiments_trump
```

```python
#creating a function to clean the tweet_texts off any links if present
#find out the values of sentiment object, sentiment value
#put them into a dataframe for further plotting

def sentiment_anl(tweetstr):
    #replcing links
    x=tweetstr.replace(r'http\S+', '',regex=True).replace(r'www\S+', '',regex=True).replace(r"\\[a-z][a-z]?[0-9]+", '',regex=True)
    
    #TextBlob is a package used for processing textual data, it helps to assign polarity values to tweets
    #creating textBlob objects
    senti_obj = [TextBlob(tweet) for tweet in x]    #https://www.earthdatascience.org/courses/earth-analytics-python/using-apis-natural-language-processing-twitter/analyze-tweet-sentiments-in-python/
    
    #using .polarity to find out the polarity associated with the textBlob object
    senti_val = [[tweet.sentiment.polarity, str(tweet)] for tweet in senti_obj]          
    senti_df = pd.DataFrame(senti_val, columns=["polarity", "tweet text"])
    return senti_df
```

```python
bernie_sentiments_df = sentiment_anl(tweet_sentiments_bernie)      
trump_sentiments_df = sentiment_anl(tweet_sentiments_trump)
trump_sentiments_df
```

```python
bernie_sentiments_df
```

```python
#creating a histogram that plots these sentiments

def plothist(sendf,titlestring):
    fig,ax = plt.subplots(figsize=(7, 7))   #initializing the size

    # Plot histogram of the polarity values
    #sendf.hist(bins=[-1, -0.25, 0.25, 1],ax=ax,color="black")
    sendf.hist(bins=[-1, -0.5, -0.25, 0.25, 0.5, 1],ax=ax,color="black")    #specifying the bins
    plt.title(titlestring, size=20)
    plt.show()
```

```python
plothist(trump_sentiments_df,"Sentiments from Tweets on Trump")
plothist(bernie_sentiments_df,"Sentiments from Tweets on Bernie")
```

```python
#### INSIGHTS ####
#We see that Bernie has a lot of tweets with positive polarity as compared to Trump related tweets
#Trump related tweets have more tweets with negative polarity as compared to Bernie related tweets
#Neutral tweets are also high for Bernie than for Trump
```

```python
#using geopy to geocode a my dataframe and find out location co-ordinates
import geopy

#the datafram has a lot of rows and this might hamper the entire procedure by sending a lot of geocoding requests at a time
# to the geocoding service used
#Inorder to avoid that, introducing and using a rate limiter that automatically adds delays betwen subsequent geocoding requests
# made to the geo-coding service
from geopy.extra.rate_limiter import RateLimiter

#Using OpenStreetMap's Nominatim which is a geocoding service that provides us with
# lattitudes, longitudes, altitudes
from  geopy.geocoders import Nominatim

#https://towardsdatascience.com/geocode-with-python-161ec1e62b89

#initializing Nominatim as the geocoder
geolocator = Nominatim(user_agent='geocoder')

#Initializing a min delay of one second between the calls sent to the geoservice  
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```


## Here begins the execution of the Geocoding block for location purposes
## Do not run the next three blocks since I've already got the location related details and stored it in an excel going by the name location_coordinates.csv

#### Please directly run the code after these subsequent 3 blocks.


```python
#applying geocode to acquire the location details in a new column location
#This step takes a long while because of the multiple geocoding requests that are sent to the geocoding service

# create location column 
#.apply(grocode) provides us with all the location details including the 3 points(lattitude, longitude, altitude) in a tuple
df['location'] = df['loc'].apply(geocode)             #https://towardsdatascience.com/geocode-with-python-161ec1e62b89
```

```python
#Over here, the details relating to latitude, longitude and altitude are collected and stored in point

df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)   #https://towardsdatascience.com/geocode-with-python-161ec1e62b89
```

```python
#Saving latitude, longitude and altitude obtained from above in 3 different columns

df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)     
```
